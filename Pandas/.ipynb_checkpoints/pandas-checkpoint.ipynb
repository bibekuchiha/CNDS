```python
#No changes have been made because drop function does not delete the rows from the original dataframe. 
#It creates a copy of the dataframe and returns the updated copy

#To make changes in the original dataframe we need to pass a inplace=True argument. This ensures that the changes are made 
#in the original dataset

iris.drop(0, inplace=True)
iris.head()
```

```python
#To delete more than one columns in one go, we can pass a list of row numbers

iris.drop([4,5,8], inplace=True)
iris.head(10)
```

```python
df =  pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df.head()
```

```python
df.columns = ['sl','sw','pl','pw','flower_types']
df.head()
```

```python
#adding new column
df['add_sl_sw'] = df['sl'] + df['sw']
df.head()
```

```python
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
setosa = df[df['flower_types']=='Iris-setosa']
setosa_sw_mean = setosa['sw'].mean()
print(setosa_sw_mean)
```

```python
setosa_pl_mean = setosa['pl'].mean()
print(setosa_pl_mean)
```

```python
df['sw'].fillna(setosa_sw_mean, inplace=True)
df['pl'].fillna(setosa_pl_mean, inplace=True)
```

```python
df.head()
```

```python
df['gender'] = 'Female'
df.head()
```

```python
df.iloc[1:20,6] = 'Male'
df.head()
```

```python
def f(s):
    if s == 'Male':
        return 0
    else:
        return 1

df['sex'] = df['gender'].apply(f)
df.head()
```
